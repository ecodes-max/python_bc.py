```python

# assignment

qst = [
    ['Someone from madagascar is called a?','Malagasy'],
    ['Humans breath out carbon(iv)oxide and plabts breath out.........?','Oxygen'],
    ['What organism lives in the dead sea?','Nothing'],
    ['What year did Hitler die?','1945'],
    ['Who is Nigeria\'s chief of Army Staff','Attahiru Muhammed' ],
    ['What does chronosphotophobia means?','Fear of time and light'],
    ['Can a whale swallow a human?','No'],
    ['Is Hippopotamus milk pink or white?','Pink'],
    ['What is black and white but red all over?','Newspaper'],
    ['What does cows drink?','Water'],
    ['Does watermelon grow from trees?','No']
]

name = input('Enter your name\n')
c_age = int(input('\nEnter your age\n'))

#code for mental age below
from random import randrange

score = 0
count = 1
end = 11

while count <=5:
    ind = randrange(0,end)
    your_qst = qst[ind][0]
    the_ans = qst[ind][1]
    
    ask = input(f'{your_qst}\n')
    
    #grade and store score
    if ask == the_ans:
        score += 5
    else:
        score += 0
        
        
    #remove question that was asked
    qst.remove(qst[ind])
    end -= 1
    
    #increment count
    count += 1
    
iq = (score/c_age) * 100

print(f'Hey {name},your IQ is {iq}')
if iq>= 80 and iq <= 109:
    print('smart dude ')
elif iq>=110 and iq <= 129:
    print('you are a genius')
elif iq>=130 and iq <=150:
    print('youre a genius')
elif iq>150:
    print('you must be einstein') 
else:
    print('your still alive')
    

#less than 80: well, you are still alive
#80-109: smart dude
#110-129: genius
#130-150: legend
#greater than 150: you must be a twin of Einstein


```
